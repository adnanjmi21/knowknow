# visualize histories

This figure shows the citation histories for the 22 citations which 
    had at least one citation per year average before death, and 
        at least five citations in the five years after rebirth.
The three dots on the x-axis represent 1900, 1950, and 2000.
Each dash represents 10 years.

```python
big_rebirth_names = cb[(cb.bef>1)&(cb.dlen>=10)].get("name")

print("NUM_REBIRTHS:", len(big_rebirth_names))

plot_count_series( 
    sorted( big_rebirth_names, key=lambda x:cysum[x]['pub'] ), 
    database=database_name,
    myname="%s.rebirth_viz" % database_name, 
    overwrite=True,
    markers={
        rb.get("name"): [rb.get("d1"),rb.get("rb")]
        for ri, rb in cb.iterrows()
    },
    cysum=cysum
)
```

# overall patterns

```python
YMAX=2
XMAX=20

def myscatter(*args, **kwargs):
    kwargs['s'] = 0.5
    kwargs['color'] = 'black'
    kwargs['alpha'] = 1
    
    filt = np.logical_or( 
        args[0] >= XMAX,
        args[1] >= YMAX
    )
    
    args = (
        args[0][filt],
        args[1][filt]
    )
    plt.scatter(*args, **kwargs)
    plt.vlines(XMAX,YMAX,plt.ylim()[0], linestyle='--', color='gray')
    plt.hlines(YMAX,plt.xlim()[0],XMAX, linestyle='--', color='gray')

g = sns.jointplot('dlen','rat',cb, kind="kde", space=0, color="g")
g.plot_joint(myscatter)

plt.xlabel("Length of death (years)")
plt.ylabel("Post-rebirth yearly citations / pre-death")
g.fig.suptitle('Length of death and popularity after rebirth');

plt.xlim(plt.xlim()[0],50)
plt.ylim(plt.ylim()[0],10)

save_figure("%s.deathLengthAndRebirthSize" % database_name)
```
