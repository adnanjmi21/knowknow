# User settings

```python
database_name = "sociology-wos"
```

```python
deathks = ["death1","death2","death3","death5"]
```

# general imports

```python
cits = get_cnt("%s.ind"%database_name, ['c.fy','fy'])
cysum = load_variable("%s.cysum"%database_name)
```

```python
from random import shuffle
# get a BigGuy
thresh = np.quantile([x['total'] for x in cysum.values()], 0.95)
bgs = [
    x for x in cysum.values() 
    if x['total'] > thresh and 
        all(x[deathk] is not None
            for deathk in deathks)
]
shuffle(bgs)
```

```python
names = [x['name'] for x in bgs]
names = sorted( names, key=lambda x:cysum[x]['pub'] )
```

```python
for deathk in deathks:
    display(HTML("<h1>%s</h1>" % deathk))
    plot_count_series( 
        names[:40],
        database=database_name,
        myname="%s.cited histories with %s labeled" % (database_name,deathk), 
        overwrite=True,
        markers={
            n: [cysum[n][deathk]]
            for n in names
        },
        cysum=cysum
    )
```

```python

display(HTML("<h1>all deaths together</h1>"))
plot_count_series( 
    names[:40],
    database=database_name,
    myname="%s.citation death1235" % (database_name), 
    overwrite=True,
    markers={
        n: [cysum[n][deathk] for deathk in deathks]
        for n in names
    },
    cysum=cysum
)
```

```python
def fix_name(x):
    sp = x.split("|")
    n, y = sp[:-1], sp[-1].strip()
    n = ", ".join( x[0].upper() + x[1:] for x in n )
    return "%s (%s)" % (n,y)

from random import random as rnd

linestyles = ['-', '--', '-.', ':']

for bg in bgs[:5]:
    
    years = range(bg['first'], 2020)
    prop_trend = [ cits['c.fy'][(bg['name'],y)] / cits['fy'][y] for y in years ]
    count_trend = [ cits['c.fy'][(bg['name'],y)] for y in years ]
    plt.plot(
        years,
        count_trend,
        color='black'
    )
    plt.fill_between(
        years,
        [ 0 for y in years ],
        count_trend,
        alpha=0.2
    )

    for i,deathk in enumerate(deathks):
        plt.axvline(bg[deathk], 0, 10, label=deathk, color="black", linestyle=linestyles[i])

    plt.title("Dates of death for %s" % fix_name(bg['name']))

    plt.legend();
    plt.xlabel("Year")
    plt.ylabel("Proportion of citations in that year")
    
    save_figure("%s.cysum_example_citationdeath" % database_name)
    plt.show();
```

```python
deathks=['death1','death2','death3']
```

```python
aysum = load_variable("%s.aysum" % database_name)
docs = get_cnt("%s.doc"%database_name, ['a.fy'])
```

```python
from random import shuffle
# get a BigGuy
thresh = np.quantile([x['total'] for x in aysum.values()], 0.95)
bgs = [
    x for x in aysum.values() 
    if x['total'] > thresh and 
        all(x[deathk] is not None 
    for deathk in deathks)
]
shuffle(bgs)
```

```python
names = [x['name'] for x in bgs]
names = sorted( names, key=lambda x:aysum[x]['first'] )
```

```python
display(HTML("<h1>author deaths together</h1>"))
plot_count_series( 
    names[:40],
    database=database_name,
    myname="%s.death123 of authors" % (database_name), 
    overwrite=True,
    markers={
        n: [aysum[n][deathk] for deathk in deathks]
        for n in names
    },
    cysum=aysum,
    ctype='a'
)
```
